```python
from wikidataloader import WikidataQuery

# Linguists from Germany with birth places and gender

results = WikidataQuery.search(
    filters={"P31": "Q5", "P27": "Q183", "P106": "Q14467526"}, # {is_instance:human, country_of_origin:Germany, profession:linguist}
    select=[("P21", "Gender"), ("P19", "City of Birth")],
    limit=15,
    default_language="en"
).to_pandas()

results
```

```python
from wikidataloader import WikidataQuery

# Images and temperatures of stars

results = WikidataQuery.search(
    filters={"P31": "Q523"}, # {is_instance:star}
    select=[("P6879", "Temperature"), ("P18", "image_url")],
    required_properties={"P18", "P6879"},
    limit=15,
    default_language="[AUTO_LANGUAGE]"
).to_pandas()

results
```

```python
from wikidataloader import WikidataQuery

# Audio pronounciations of given names

results = WikidataQuery.search(
    filters={"P31": "Q202444"}, # {is_instance:given_name}
    select=[("P407", "Language of origin"), ("P443", "audio_url")],
    required_properties={"P443"},
    limit=15,
    default_language="en"
).to_pandas()

results
```

```python
from wikidataloader import WikidataQuery

# Capital cities and languages of African countries with coordinates

results = WikidataQuery.search(
    filters={"P31": "Q3624078", "P30": "Q15"}, # {is_instance:country, continent:Africa}
    negative_filters={"P31": "Q3024240", "P31": "Q28171280"}, # not: {is_instance: historical_country, is_instance:ancient_civilization}
    select=[("P36", "Capital"), ("P37", "Language"), ("P625", "Coordinates")],
    limit=15,
    default_language="en"
).to_pandas()

results
```

```python
from wikidataloader import WikidataQuery

# Color terms with lexicographical information

results = WikidataQuery.search(
    filters={"P31": "Q376431"}, # {is_instance:color_term}
    retrieve_lexicographical_information=True,
    limit=11,
    default_language="en"
).to_pandas()

results
```
